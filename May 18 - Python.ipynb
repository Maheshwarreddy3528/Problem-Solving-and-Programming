## Problem 3 :

### Problem Statement 
Plot a line graph on **income vs year** for the state of California
### Constraints 

### Test Cases
*

```python
import pandas as pd
df = pd.read_csv('DataFiles/Income.csv')
y = df.values[4]
x = df.columns
x
y
```

```python
years = []
for i in range(2,len(x)):
    years.append(x[i])
years
income=[]
for i in range(2,len(y)):
    income.append(y[i])
income
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years,income)
plt.title('Income of California state')
plt.grid(True)
plt.savefig('Figures/Year-Income.jpeg')
plt.show()
```
